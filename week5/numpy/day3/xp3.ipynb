```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create a line plot and scatter plot representing your favorite mathematical functions. 
# Change the line color to tab:blue for the line plot and use a marker of your choice for the scatter plot.
figure, axes = plt.subplots(2, sharex=True)
axes[0].set(title = 'y=sin(x)')
axes[0].plot(x, y1, color='blue')
axes[1].set(title='y=cos(x)')
axes[1].scatter(x, y2, color='green', marker='x')
```

```python
#Plot the line plot and scatter plot from Exercise 7 on the same figure using subplots. 
#Use plt.subplots to create a figure with two subplots.
figure, axes = plt.subplots(2, sharex=True)
axes[0].set(title = 'y=sin(x)')
axes[0].plot(x, y1, color='blue')
axes[1].set(title='y=cos(x)')
axes[1].scatter(x, y2, color='green', marker='x')
```

```python
figure, axes = plt.subplots(2, sharex=True)
axes[0].set(title = 'y=sin(x)')
axes[0].plot(x, y1, color='blue', label='y=sin(x)')
axes[0].legend()
axes[1].set(title='y=cos(x)')
axes[1].scatter(x, y2, color='green', marker='x', label='y=cos(x)')
axes[1].legend()
```
