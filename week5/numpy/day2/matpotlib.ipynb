```python
# Exercise 10 Advanced Plots
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Create a meshgrid
x_mesh, y_mesh = np.meshgrid(x, y)
z_mesh = y_mesh**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x_mesh, y_mesh, z_mesh, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Y^2')
ax.set_title('3D Plot of Y^2')

# Add colorbar
fig.colorbar(surf)

# Show the plot
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x)

# Generate error values
error = 0.2

# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(x, y, yerr=error, label='Sine Function', color='blue', linewidth=2)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot of Sine Function with Error Bars')
plt.legend()
```
