```python
axes_set[0]
```

```python
x = np.linspace(1,100,100)
y1 = x ** 2
y2 = x ** 3
figure, axes_set = plt.subplots(2)
axes_set[0].plot(x,y1, color='red', label='main 1 line')
axes_set[0].legend(title='Legend', shadow=True, loc='lower right')
axes_set[0].annotate('breakout!',xy=(0,0), xytext=(1,2000),arrowprops={'facecolor':'black', 'shrink':10})
axes_set[1].plot(y2, color='green')
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

figure, axes_set = plt.subplots(4)
axes_set[0].plot(x,y, color='red', label='main 1 line')
axes_set[1].plot(x, -y, color='green')
axes_set[2].plot(x, (y*2), color='black')
axes_set[3].plot(x, (-y*2), color='purple')

```

```python
max_index = np.argmax(y)
min_index = np.argmin(y)
figure, axes_set = plt.subplots(4)
axes_set[0].plot(x,y, color='red', label='main 1 line')

axes_set[0].annotate(f'Max: ({x[max_index]:.2f}, {y[max_index]:.2f})',
             xy=(x[max_index], y[max_index]), xytext=(10, -40),
             textcoords='offset points', arrowprops=dict(facecolor='green', arrowstyle='wedge,tail_width=0.8'),
             fontsize=10, ha='center')
axes_set[0].annotate(f'Min: ({x[min_index]:.2f}, {y[min_index]:.2f})',
             xy=(x[min_index], y[min_index]), xytext=(10, 20),
             textcoords='offset points', arrowprops=dict(facecolor='red', arrowstyle='wedge,tail_width=0.7'),
             fontsize=10, ha='center')
axes_set[0].legend(title='Legend', shadow=True, loc='upper right')
axes_set[1].plot(x, -y, color='green')
axes_set[2].plot(x, (y*2), color='black')
axes_set[3].plot(x, (-y*2), color='purple')
```
