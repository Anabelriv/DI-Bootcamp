```python
boston_data = pd.read_csv("BostonHousing.csv")

# Create a DataFrame from the data and target
feature_names = list(boston_data.columns)[:-1]+['MEDV']

boston_data.columns = feature_names

boston_df = boston_data

boston_df.head()
```

```python
#missing data
missing_data = boston_df.isnull().sum()
missing_data
```

```python
#drop NA rows
boston_df = boston_df.dropna()
```

```python
# normalize data
min_max = {}
for column in boston_df.columns[:-1]:
    min_val = boston_df[column].min()
    max_val = boston_df[column].max()
    min_max[column] = {'min': min_val, 'max': max_val}
    boston_df[column] = (boston_df[column] - min_val) / (max_val - min_val)
boston_df.head()
```

```python
# Display min and max values for each column
for column, values in min_max.items():
    print(f"Column: {column}, Min: {values['min']}, Max: {values['max']}")
```

```python
# Exercise Categorical

boston_df['age_category'] = pd.cut(boston_df['age'],bins=[0.0,0.30,0.60,0.90],labels=['new','old','very_old'])
boston_df.head()
```

```python
boston_df['RM_LSTAT'] = boston_df['rm'] * boston_df['lstat']
boston_df.head()
```

### Part 2

```python
new_properties = boston_df[boston_df['age'] <= 0.35]
old_properties = boston_df[(boston_df['age'] > 0.35) & (boston_df['age'] <= 0.70)]
very_old_properties = boston_df[boston_df['age'] > 0.70]
plt.figure(figsize=(10, 6))

plt.hist(new_properties['MEDV'], bins=20, alpha=0.5, label='New Properties')
plt.hist(old_properties['MEDV'], bins=20, alpha=0.5, label='Old Properties')
plt.hist(very_old_properties['MEDV'], bins=20, alpha=0.5, label='Very Old Properties')

plt.xlabel('Median Value (MEDV)')
plt.ylabel('Frequency')
plt.title('Property Age vs. Median Value')
plt.legend()

plt.show()
```

### Part 3

```python
boston_df['CRIM_BIN'] = pd.cut(boston_df['crim'], bins=[0.0, 0.01, 0.05, 0.10, 0.20, max(boston_df['crim'])], labels=['0-1', '1-5', '5-10', '10-20', '20+'])
```

```python
avg_MEDV_per_CRIM_BIN = boston_df.groupby('CRIM_BIN')['MEDV'].mean()
```

```python
plt.figure(figsize=(10,6))

plt.plot(avg_MEDV_per_CRIM_BIN, marker='o')

plt.xlabel('Binned Crime Rate')
plt.ylabel('Average MEDV (Price)')
plt.title('Neighborhood Crime Rate Impact on Prices')
plt.show()
```
